```python
# 4. выделите леммы в значениях столбца с целями получения кредита:
# -опишите, как вы проводили лемматизацию целей кредита;
```

```python
from pymystem3 import Mystem

# получаем стеммер для слов на русском языке
m = Mystem()
```

```python
# создаем функцию для лемматизации текста
def lemmatize_text(text):
    lemmas = m.lemmatize(text)
    return ''.join(lemmas).strip()
```

```python
# выбираем уникальные значения в столбце целя получения кредита
purpose = pd.Series(df_cleaned['purpose'].unique())
purpose
```

```python
# применяем функцию леммитизации к уникальным значениям
lemmatized_purposes = purpose.apply(lemmatize_text)
```

```python
# создаем словарь для замены исходных значений на леммитизированные
lemmatization_dict = dict(zip(purpose, lemmatized_purposes))
```

```python
# применяем этот словарь к столбцу целей кредита
df_cleaned.loc[:, 'purpose_lemmas'] = df_cleaned['purpose'].map(lemmatization_dict)
```

```python
df_cleaned[['purpose', 'purpose_lemmas']].head(20)
```

```python
# лемматизация целей кредита проведена с помощью библиотеки pymystem3
# сначала я извлекла уникальные значения из столбца purpose, затем применила лемматизацию к каждому уникальному значению, 
# создав словарь для замены исходных значений на лемматизированные
```

```python
# создадим для этого отдельную таблицу
df_log = df_cleaned[['children', 'family_status_id', 'debt']]
df_log
```

```python
df_cleaned[['family_status', 'family_status_id']].value_counts()
```

```python
def family_status(status):
    if status == 0:
        return 'женат / замужем'
    elif status == 1:
        return 'гражданский брак'
    elif status == 2:
        return 'вдовец / вдова'
    elif status == 3:
        return 'в разводе'
    elif status == 4:
        return 'не женат / не замужем'
    
# применяем к столбцу датафрейма нашу функцию
df_log.loc[:, 'family_status_id'] = df_log['family_status_id'].apply(family_status)
```

```python
df_log['children'].value_counts()
```

```python
# видим аномальные значения (артефакты) в виде 20 детей, исправим это 
# заменяем 20 на 2
df_log.loc[:, 'children'] = df_log['children'].replace(20, 2)

# проверим, как теперь распределены данные
df_log['children'].value_counts()
```

```python
# функция, которая категоризирует количество детей 
def categorize_children(children):
    if children == 0:
        return 'нет детей'
    elif children == 1:
        return '1 ребёнок'
    elif children == 2:
        return '2 детей'
    elif children >= 3:
        return '3 и больше детей'
    
# применяем к столбцу датафрейма нашу функцию
df_log.loc[:, 'children'] = df_log['children'].apply(categorize_children)
```

```python
# группируем по семейному положению и считаем среднюю задолженность
family_status_analysis = df_log.groupby('family_status_id')['debt'].mean().reset_index()

# группируем по количеству детей и считаем среднюю задолженность
children_analysis = df_log.groupby('children')['debt'].mean().reset_index()
```

```python
# я выделила следующие «словари» и категории, которые помогают структурировать данные и упростить их дальнейший анализ:
# 1) семейное положение
# категоризация семейного положения позволяет изучить, как различные семейные обстоятельства 
# влияют на платежеспособность клиента
# 2) количество детей
# поскольку количество детей может существенно влиять на финансовую нагрузку семьи, 
# важно создать категории для удобного анализа
```

```python
# создадим еще одну таблицу
df_dop = df_cleaned[['total_income', 'purpose_lemmas', 'debt']]
df_dop
```

```python
# посмотрим какие доходы имеются
print(df_dop['total_income'].value_counts())
print('Минимум', df_dop['total_income'].min())
print('Максимум', df_dop['total_income'].max())
```

```python
# создаем функцию, которая категоризирует количество ежемесячного дохода 
def categorize_total_income(income):
    if income <= 70000:
        return 'до 70к'
    elif income <= 100000:
        return '70к-100к'
    elif income <= 200000:
        return '100к-200к'
    elif income <= 500000:
        return '200к-500к'
    else:
        return 'больше 500к'
    
# применяем к столбцу датафрейма нашу функцию
df_dop.loc[:, 'total_income'] = df_dop['total_income'].apply(categorize_total_income)
df_dop['total_income'].value_counts()
```

```python
# группируем по ежемесячному доходу и подсчитываем среднюю задолженность
total_income_analysis = df_dop.groupby('total_income')['debt'].mean().reset_index()
```

```python
# проводим подсчет слов целей кредита и видим, что цели еще не категоризированы
from collections import Counter
Counter(df_dop['purpose_lemmas'])
```

```python
# создаем функцию для категоризации данных по целям кредита
def categorize_purpose(purpose):
    if 'образование' in purpose:
        return 'образование'
    elif 'автомобиль' in purpose:
        return 'автомобиль'
    elif 'свадьба' in purpose:
        return 'свадьба'
    elif 'недвижимость' in purpose:
        return 'недвижимость'
    elif 'жилье' in purpose:
        return 'жилье'
    else:
        return 'другое'

# применяем к столбцу датафрейма нашу функцию
df_dop.loc[:, 'purpose_lemmas'] = df_dop['purpose_lemmas'].apply(categorize_purpose)
df_dop['purpose_lemmas'].value_counts()
```

```python
# группируем по целям получения кредита и подсчитываем среднюю задолженность
purpose_analysis = df_dop.groupby('purpose_lemmas')['debt'].mean().reset_index()
```
